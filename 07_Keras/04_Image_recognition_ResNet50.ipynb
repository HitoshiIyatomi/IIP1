### One-page example: ResNet50 image classification

Extract essential part of the classification

```python
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np


# download and define pre-trained model
model = ResNet50(weights='imagenet')
```

Upload your test image to your Colab root  

in this example,  'pics.jpg'


```python
import matplotlib.pyplot as plt

# set target image (you should prepare image file to local)
img_path = 'pics.jpg'

img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
print(x.shape)
x = np.expand_dims(x, axis=0)
print(x.shape)
x = preprocess_input(x)

print(x.shape)
plt.imshow(np.squeeze(x))

preds = model.predict(x)

#plt.imshow(img)
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
print('Predicted:', decode_predictions(preds, top=5)[0])
# Predicted: [(u'n02504013', u'Indian_elephant', 0.82658225), (u'n01871265', u'tusker', 0.1122357), (u'n02504458', u'African_elephant', 0.061040461)]
```

```python
# set target image (you should prepare image file to local)
img_path = 'pics.jpg'

img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
#print(x.shape)
x = np.expand_dims(x, axis=0)
#print(x.shape)
x = preprocess_input(x)
preds = model.predict(x)

plt.figure(figsize=(8, 16))
ax = plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("input image")
ax = plt.subplot(1, 2, 2)
plt.imshow(np.squeeze(x))
plt.title("pre-processed")


#plt.imshow(img)
# decode the results into a list of tuples (class, description, probability)
print('Predicted:', decode_predictions(preds, top=5)[0])
# (ex) Predicted: [(u'n02504013', u'Indian_elephant', 0.82658225), (u'n01871265', u'tusker', 0.1122357), (u'n02504458', u'African_elephant', 0.061040461)]
```
